###3.혼인건수와 출생아수 데이터 재정리하기

```python
#열이름 변경하기
#변수명.rename(columns = {‘열이름’:‘새로운 열이름’}, inplace= True)
#출생아수(명)->출생아수
#혼인건수(건)->혼인건수

data.rename(columns={'출생아수(명)':'출생아수', '혼인건수(건)':'혼인건수'}, inplace=True)
data
```

```python
#2010년도~2021년도 데이터 행 찾기

data[10:]
```

```python
#2010년도~2021년도 데이터 행 저장하기

numData = data[10:]
numData
```

```python
# 2010년이 10번째 행임으로 index 정리하고 확인하기
#변수명.reset_index(drop=True, inplace=True)

numData.reset_index(drop=True, inplace=True)
numData.head()
```

####2)차트를 생성하기 위한 라이브러리 읽어오기


```python
import matplotlib.pyplot as plt
```

####3) 혼인건수가 가장 적은 년도와 가장 많은 년도는 언제인가?


```python
#년도에 따른 혼인건수를 plot차트로 생성하기
#plt.plot(y축데이터)
#plt.show()

plt.plot(numData['혼인건수'])
plt.show()
```

```python
#x축의 눈금값이 모두 나타나도록 차트 수정하기

#plt.xticks(range(눈금값), 레이블)
#- 눈금값: x, y 축 눈금에 표시할 눈금(range(), numpy.arrange())
#- 레이블: 눈금값으로 표시할 레이블로 튜플이나 리스트로 설정

plt.plot(numData['혼인건수'])
plt.xticks(range(len(numData['시점'])), numData['시점'])
plt.show()
```

####4) 출생아수가 가장 적은 년도와 가장 많은 년도는 언제인가?


```python
#년도에 따른 혼인건수를 plot차트로 생성하기
#plt.plot(y축데이터)
#y축데이터: 출생아수

plt.plot(numData['출생아수'])
plt.xticks(range(len(numData['시점'])), numData['시점'])
plt.show()
```

####5) 출생아수가 혼인건수에 영향이 있는가?

```python
##년도에 따른 혼인건수와 출생아수를 하나의 plot차트로 생성하기(범례 삽입)
#plt.plot(y축데이터, label='범례이름')
# label='범례이름': plot 차트 선의 이름
#plt.legend(): 범례를 차트내에 표시

plt.plot(numData['혼인건수'], label='혼인건수')
plt.plot(numData['출생아수'], label='출생아수')

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])

plt.show()
```

####6) 차트 꾸며보기

```python
#모든 차트에 marker 추가하고 '출생아수' 차트의 markersize 조절하기
#marker='마커종류': 선 위에 표현할 마커 모양 설정
#markersize=정수: 마커의 크기 변경

plt.plot(numData['혼인건수'], label='혼인건수', marker='.')
plt.plot(numData['출생아수'], label='출생아수', marker='*', markersize='10')

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])

plt.show()
```

```python
#모든 차트에 linestyle 추가하고 출생아수 차트의 선두께 조절하기
#linestyle='선종류': 선의 모양 설정
#linewidth=정수: 선 두께 설정

plt.plot(numData['혼인건수'], label='혼인건수', marker='.', linestyle='--')
plt.plot(numData['출생아수'], label='출생아수', marker='*', markersize='10', linestyle=':', linewidth=3)

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])

plt.show()
```

```python
#차트 색상 변경하기
#color='색상': 선 색상 설정

plt.plot(numData['혼인건수'], label='혼인건수', marker='.', linestyle='--', color='y')
plt.plot(numData['출생아수'], label='출생아수', marker='*', markersize='10', linestyle=':', linewidth=3, color='g')

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])

plt.show()
```

```python
#혼인건수 차트의 색상, 마커, 선종류를 약식으로 표현해보기
#'ro:' : color='r', marker='o', linestyle='-'

plt.plot(numData['혼인건수'], 'y.--', label='혼인건수') # plot과 label 사이에 color,marker,linestyle 순서로
plt.plot(numData['출생아수'], 'g*:', label='출생아수', markersize='10', linewidth=3)

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])

plt.show()
```

```python
#차트제목, XY 레이블, grid 지정해보기
#plt.title('차트제목') :지정한 차트제목을 차트 중앙 상단에 표시
#plt.xlabel('x축제목'), plt.ylabel('y축제목') : x축과 y축의 제목을 지정한 문자열로 설정
#plt.ylim(처음값, 마지막값)
#plt.grid() : 차트내에 눈금선 표시

plt.plot(numData['혼인건수'], 'y.--', label='혼인건수') # plot과 label 사이에 color,marker,linestyle 순서로
plt.plot(numData['출생아수'], 'g*:', label='출생아수', markersize='10', linewidth=3)

plt.legend()
plt.xticks(range(len(numData['시점'])), numData['시점'])
plt.title("2010~2021년 혼인건수에 따른 출생아수")
plt.xlabel("년도")
plt.ylabel("혼인건수(수) & 출생아수(명)")
plt.ylim(150000, 500000)
plt.grid()

plt.show()
```
