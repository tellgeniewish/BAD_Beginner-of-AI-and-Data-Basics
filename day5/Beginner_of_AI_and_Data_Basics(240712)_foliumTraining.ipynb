```python
import folium

folium.Map()
```

```python
map = folium.Map(location=[37.6071, 127.0424],
                 zoom_start=15,
                 zoom_control=False,
                 control_scale=True)
map
```

```python
mk = folium.Marker(location=[37.6071, 127.0424]).add_to(map)
map
```

```python
mk = folium.Marker(location=[37.6071, 127.0424], popup="이곳은 학교") # 위치를 클릭하면 메세지가 뜬다
mk.add_to(map)
map
```

```python
# 메세지를 가로로 띄우고 싶으면
mk = folium.Marker(location=[37.6071, 127.0424], popup=folium.Popup("이곳은 학교", max_width=100))
mk.add_to(map)
map
```

```python
mk2 = folium.Marker([37.6071, 127.045], tooltip="위치2") # 클릭하지 않고 마우스만 가져가도 메세지가 뜬다
mk2.add_to(map)

map
```

```python
mk3 = folium.Marker([37.6071, 127.041], tooltip="위치3",
                    icon=folium.Icon()) # 다른 아이콘
mk3.add_to(map)

map
```

```python
mk4 = folium.Marker([37.606, 127.043], tooltip="위치4",
                    # 아이콘 그림과 색상 지정 ㄱㄴ
                    icon=folium.Icon(color='red',
                                     icon='hippo', # https://fontawesome.com/search?s=thin&o=r 에 있는 아이콘 그림을 icon에 사용 ㄱㄴ
                                     prefix='fa')) # 외부에서 아이콘을 가져와서 사용하면 fa
mk4.add_to(map)

map
```

```python
mk5 = folium.Circle([37.606, 127.042], tooltip="위치5",
                    color='green')
mk5.add_to(map)

map
```

```python
mk6 = folium.Circle([37.606, 127.042], tooltip="위치6",
                    color='orange', radius=100)
mk6.add_to(map)

map
```

```python
mk7 = folium.Circle([37.605, 127.041], tooltip="위치7",
                    color='yellow', radius=300, fill=True)
mk7.add_to(map)

map
```
